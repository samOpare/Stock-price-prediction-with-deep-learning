```python
%matplotlib inline
from model import FFNN
from dataloader import DataLoader
from evaluator import Evaluator
import pandas as pd
from feature_selection import feature_selection
from indicators import stochastic_oscillator
from indicators import larry_william_r
ffnn = FFNN()
evaluator = Evaluator()
data_loader = DataLoader()
```

# Train on all data without shuffling

```python
train_data, test_data = ffnn.get_train_test(data_with_apple_target, shuffle=False)
ffnn.fit(train_data.drop(['Target'], axis=1), train_data['Target'])
```
